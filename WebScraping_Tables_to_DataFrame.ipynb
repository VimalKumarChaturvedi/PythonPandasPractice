```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

def  get_table_from_url(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page, 'html5lib')
    rows = []  #List of Table rows
    
    for child_rows_of_table in soup.find_all('table')[3].find_all(name = 'tr'):   
        row_data = []
        for table_header in child_rows_of_table.find_all(name = ['td', 'th']):
            row_data.append(table_header.text.replace('\n', ''))
        rows.append(row_data)
    header = rows[0]
    #There is an issue with the data . The data rows have extra <td> at start which header rows does not have
    #To make header rows and data rows to have same number of elements , add one extra element in header at start 
    # That will be latter dropped 
    # Other solution is in the loop check if first element is null then drop that element 
    header.insert(0, 'tobedropped')
    df = pd.DataFrame(rows[1:], columns =  header)
    return df      
     

```

```python
url = "https://en.wikipedia.org/wiki/Indian_Premier_League"    
table = get_table_from_url(url)
table.columns
table.drop('tobedropped', axis =1, inplace = True )
table.replace('', np.nan, inplace = True)
table.dropna(subset= ['Team'] , inplace =  True)
table
```
